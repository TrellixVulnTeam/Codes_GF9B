```python
import Augmentor
import os
from PIL import Image
import random
```

```python
path = '/home/ayon/Pictures/Pics'
p = Augmentor.Pipeline(source_directory = path, output_directory = path)
p.rotate(probability = 1.0, max_left_rotation = 20, max_right_rotation = 20)
p.shear(probability = 0.6, max_shear_left = 15, max_shear_right = 15)
# p.flip_random(probability = 0.8)
# p.skew(probability=0.6, magnitude=1)
# p.random_erasing(probability = 0.8, rectangle_area = 0.3)
p.sample(100)
```

```python
root = '/home/ayon/Pictures/Pics'
img_paths = [os.path.join(root, file) for root, dirs, files in os.walk(root) for file in files]
im = [Image.open(img_path) for img_path in img_paths]
im_resize = [i.resize((256,256)) for i in im]
im_sam = random.sample(im_resize, 4)
len(im_sam)
for r in range(200):
    im_bg  = Image.open('/home/ayon/Pictures/*.JPEG')
    k,c = (0,0)
    im_sam = random.sample(im_resize, 4)
    for i in range(len(im_sam)):
        print((k,c)),print(i)
        im_bg.paste(im_sam[i],(k,c))
        if(i%2 != 0):
            k = 0
            c = 256
        elif(i>0 and i%2 == 0): 
            c = 0
            k = 256
        else: 
            c = 256
            k = 256
        im_bg.paste(im_sam[i],(k,c))
        im_bg.load()
    im_sam = im_sam[random.randint(0,3)].convert('RGBA')
    im_sam = im_sam.rotate(random.randint(30,46), expand=0).resize((random.randint(300,512),random.randint(300,512)))
    im_bg.paste(im_sam,(random.randint(0,128),random.randint(0,128)),im_sam)
    im_bg.load()
    im_bg.save('/home/ayon/Pictures/Gen_Scrambled/New%d.png'%r)
    im_bg.close()
```
